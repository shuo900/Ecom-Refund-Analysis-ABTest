```python
import pandas as pd
import io

# 1. 定义文件名映射（请确保你的文件夹里文件名和下面一致）
# 如果你已经重命名了文件，请修改下面的字符串
file_map = {
    'sales': 'sales_detail_generated.csv',
    'refund': 'refund_detail_generated.csv',
    'chat': 'beauty.xlsx'
}

def smart_read(file_name):
    for enc in ['utf-8', 'gbk', 'utf-8-sig']:
        try:
            df = pd.read_csv(file_name, encoding=enc)
            print(f"✅ 成功读取 {file_name} (使用编码: {enc})")
            return df
        except:
            continue
    print(f"❌ 无法读取 {file_name}，请检查文件是否存在")
    return None

# 执行读取
df_sales = smart_read(file_map['sales'])
df_refund = smart_read(file_map['refund'])
df_chat = smart_read(file_map['chat'])

# 2. 预处理：统一时间格式（这对后续统计非常重要）
if df_sales is not None:
    df_sales['order_time'] = pd.to_datetime(df_sales['order_time'])
if df_refund is not None:
    df_refund['order_time'] = pd.to_datetime(df_refund['order_time'])
    df_refund['refund_time'] = pd.to_datetime(df_refund['refund_time'])
if df_chat is not None:
    df_chat['inquiry_time'] = pd.to_datetime(df_chat['inquiry_time'])

print("\n--- 所有数据已就绪 ---")
```

```python
# 1. 基础清洗：去除 ID 两端的空格
df_sales['order_id'] = df_sales['order_id'].astype(str).str.strip()
df_refund['order_id'] = df_refund['order_id'].astype(str).str.strip()

# 2. 统计学思维：左连接（Left Join）销售与退款表
# 我们以销售表为准，看看每笔订单是否发生了退款
df_merged = pd.merge(df_sales, 
                     df_refund[['order_id', 'refund_time', 'refund_quantity', 'refund_reason']], 
                     on='order_id', 
                     how='left')

# 3. 标记是否退款：如果 refund_time 是空的，说明没退款
df_merged['is_refund'] = df_merged['refund_time'].notnull().astype(int)

# 4. 预览合并后的关键信息
print("--- 订单-退款合并表概览 ---")
display(df_merged[['order_id', 'quantity', 'avg_actual_unit_price', 'is_refund']].head())
```

```python
# 1. 划分实验组与对照组
df_merged['group'] = df_merged['quantity'].apply(lambda x: '异常组(3件)' if x == 3 else '正常组(1-2件)')

# 2. 计算各组的退款表现
comparison = df_merged.groupby('group').agg(
    订单总数=('order_id', 'count'),
    退款总数=('is_refund', 'sum'),
    平均单价=('avg_actual_unit_price', 'mean')
).reset_index()

comparison['退款率'] = comparison['退款总数'] / comparison['订单总数']

print("--- 📊 核心归因分析：为什么 3 件组退款率高？ ---")
display(comparison)

# 3. 统计学思考：计算退款风险倍数 (Relative Risk)
risk_ratio = comparison.loc[1, '退款率'] / comparison.loc[0, '退款率']
print(f"\n💡 结论：购买3件的退款风险是普通订单的 {risk_ratio:.2f} 倍！")
```

```python
# 1. 修正风险倍数计算 (确保分子是异常组)
risk_abnormal = comparison[comparison['group'] == '异常组(3件)']['退款率'].values[0]
risk_normal = comparison[comparison['group'] == '正常组(1-2件)']['退款率'].values[0]
actual_risk_ratio = risk_abnormal / risk_normal

print(f"✅ 修正后的结论：购买3件的退款风险是普通订单的 {actual_risk_ratio:.2f} 倍！")

# 2. 客服语料文本挖掘：提取“价格敏感”型吐槽
keywords = ['20', '80', '算错', '价格', '贵', '不符', '差价']
# 筛选包含关键词的聊天记录
df_chat['is_price_complaint'] = df_chat['content'].str.contains('|'.join(keywords), na=False)

# 看看这些吐槽都分布在什么类别里
complaint_summary = df_chat[df_chat['is_price_complaint']].groupby('category').size().sort_values(ascending=False)
print("\n--- 📢 价格相关客诉的分布类别 ---")
print(complaint_summary)
```

```python
# 导出清洗后的数据，方便导入 SQL
# index=False 是为了不把 Pandas 的行索引带进数据库
df_sales.to_csv('sales_for_sql.csv', index=False, encoding='utf-8')
df_refund.to_csv('refund_for_sql.csv', index=False, encoding='utf-8')
df_chat.to_csv('chat_for_sql.csv', index=False, encoding='utf-8')

print("✅ 导出完成！请将这三个文件导入你的 MySQL/Navicat 中。")
```

```python
import numpy as np
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
from statsmodels.stats.power import NormalIndPower

# 1. 实验前方案设计：计算最小样本量 (Sample Size Calculation)
# 假设基准转化率为 10%，我们希望检测到 2% 的绝对提升 (MDE=2%)
effect_size = 0.12 - 0.10 
# 使用 statsmodels 计算样本量
analysis = NormalIndPower()
required_n = analysis.solve_power(effect_size=0.1, alpha=0.05, power=0.8, ratio=1.0)
print(f"根据统计功效要求，每组建议最小样本量: {int(np.ceil(required_n))}")

# 2. 模拟实验数据 (假设实验已结束)
# A组 (对照组)：原复杂价格页面； B组 (实验组)：直降后到手价页面
data = {
    'group': ['A', 'B'],
    'conversions': [420, 510],  # 下单人数
    'total_users': [4000, 4050] # 进店总人数
}
df_ab = pd.DataFrame(data)

# 3. 显著性检验 (Z-test for Proportions)
count = np.array(df_ab['conversions'])
nobs = np.array(df_ab['total_users'])

z_stat, p_value = proportions_ztest(count, nobs)

# 4. 计算置信区间
conv_a = 420/4000
conv_b = 510/4050
lift = (conv_b - conv_a) / conv_a

print(f"--- 实验结果 ---")
print(f"A组转化率: {conv_a:.2%}, B组转化率: {conv_b:.2%}")
print(f"相对提升 (Lift): {lift:.2%}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("结论：在 95% 置信水平下，实验组结果显著优于对照组，建议全量上线该方案。")
else:
    print("结论：结果不显著，可能需要延长实验时长或重新审视策略。")
```
